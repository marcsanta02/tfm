```python
dt_cols = [
    'dt_last_access_date', 'dt_second_last_access_date',
    'dt_third_last_access_date', 'dt_fourth_last_access_date',
    'dt_fifth_last_access_date', 'lifetime'
]
for col in dt_cols:
    df[f'log_{col}'] = np.log1p(df[col])
df.drop(columns=dt_cols, inplace=True, errors="ignore")
access_cols = [
    'access_count_last_1_day', 'access_count_last_3_days',
    'access_count_last_7_days', 'access_count_last_15_days'
]
df[access_cols] = df[access_cols].fillna(0)

df['ratio_1d_15d'] = df['access_count_last_1_day'] / (df['access_count_last_15_days'] + 1)
df['ratio_3d_15d'] = df['access_count_last_3_days'] / (df['access_count_last_15_days'] + 1)
df['ratio_1d_3d'] = df['access_count_last_1_day'] / (df['access_count_last_3_days'] + 1)
df['ratio_3d_7d'] = df['access_count_last_3_days'] / (df['access_count_last_7_days'] + 1)

df['access_trend'] = (
    df['access_count_last_1_day'] - df['access_count_last_3_days']/3 +
    df['access_count_last_3_days'] - df['access_count_last_7_days']/7
)
# Ordered mapping (ensure the order is consistent with your definitions)
temperature_map = {'cold': 0, 'cold-warm': 1, 'warm-hot': 2, 'hot': 3}
size_category_map = {'small': 0, 'medium': 1, 'large': 2, 'xlarge': 3}

df['temperature_encoded'] = df['temperature'].map(temperature_map)
df['size_category_encoded'] = df['size_category'].map(size_category_map)

df.drop(columns=['temperature', 'size_category'], inplace=True, errors='ignore')
df['type_binary'] = (df['type'] == 'mc').astype(int)
df.drop(columns='type', inplace=True, errors='ignore')
```

```python
def split_dataframes(start, df=df):
    train_window_size = 13
    end = start + train_window_size

    df_train = df[(df['period'] <= end) & (df['period'] >= start)].copy()
    df_val = df[df['period'] == end].copy()
    df_test = df[df['period'] == end + 1].copy()
    
    period_range = end - start
    df_train['period_norm'] = (df_train['period'] - start) / period_range
    df_val['period_norm'] = (df_val['period'] - start) / period_range
    df_test['period_norm'] = (df_test['period'] - start) / period_range

    # Save test file_ids before dropping
    test_file_ids = df_test[['pnfsid', 'size', 'y1']].reset_index(drop=True)

    # Drop columns not used for features
    X_train, y_train = df_train.drop(columns=['y1', 'period', 'pnfsid', 'size']), df_train['y1']
    X_val, y_val = df_val.drop(columns=['y1', 'period', 'pnfsid', 'size']), df_val['y1']
    X_test, y_test = df_test.drop(columns=['y1', 'period', 'pnfsid', 'size']), df_test['y1']

    return X_train, y_train, X_val, y_val, X_test, y_test, test_file_ids
```

```python
def obtain_predictions(X_train, y_train, X_val, y_val, X_test, y_test, best_params):
    # Undersampling
    rus = RandomUnderSampler(sampling_strategy=1.0, random_state=42)
    X_train_resampled, y_train_resampled = rus.fit_resample(X_train, y_train)

    # DMatrix for XGBoost
    dtrain = xgb.DMatrix(X_train_resampled, label=y_train_resampled, enable_categorical=True)
    dval = xgb.DMatrix(X_val, label=y_val, enable_categorical=True)
    dtest = xgb.DMatrix(X_test, label=y_test, enable_categorical=True)

    # Model parameters
    params = {
        'objective': 'binary:logistic',
        'max_depth': best_params['max_depth'],
        'learning_rate': best_params['learning_rate'],
        'subsample': best_params['subsample'],
        'colsample_bytree': best_params['colsample_bytree'],
        'min_child_weight': best_params['min_child_weight'],
        'gamma': best_params['gamma'],
        'reg_alpha': best_params['reg_alpha'],
        'reg_lambda': best_params['reg_lambda'],
    }

    early_stop = EarlyStopping(rounds=best_params['early_stopping_rounds'], min_delta=best_params['min_delta'], save_best=True)
    model = xgb.train(
        params,
        dtrain,
        num_boost_round=1000,
        evals=[(dtrain, 'train'), (dval, 'eval')],
        callbacks=[early_stop],
        verbose_eval=False
    )

    # Predict
    return model.predict(dtest)
```

```python
def lru_predictions(X_test):
    """
    Simulate LRU predictions based on recent access patterns.
    Files with more recent access get higher prediction scores.
    """
    lru_scores = np.zeros(len(X_test))
    
    max_val = X_test['log_dt_last_access_date'].max()
    
    lru_scores = (max_val - X_test['log_dt_last_access_date'])
    
    lru_scores = (lru_scores - lru_scores.min()) / (lru_scores.max() - lru_scores.min())
    
    return lru_scores
```

```python
all_results = []  # List of DataFrames (one per period)
max_i = df['period'].max() - 14

# Initial best params as we start at i 115
with open("../Hyperparams/v2_best_params_period_112_125.json", 'r') as file:
    best_params = json.load(file)   

for i in range(115, max_i, 1):
    if i % 14 == 0:
        if i <= 182:
            filename = f"../Hyperparams/v2_best_params_period_{i}_{i+13}.json"
        else: 
            filename = f"../Hyperparams/v2_best_params_period_182_195.json"
        with open(filename, 'r') as file:
            best_params = json.load(file)       
    
    X_train, y_train, X_val, y_val, X_test, y_test, test_file_ids = split_dataframes(i, df)
    y_pred_test_proba = obtain_predictions(X_train, y_train, X_val, y_val, X_test, y_test, best_params)

    test_file_ids['prediction'] = y_pred_test_proba
    test_file_ids['period'] = i

    test_file_ids = test_file_ids.sort_values('prediction', ascending=False).reset_index(drop=True)

    all_results.append(test_file_ids)
```

```python
def simulate_cache(
    all_results,
    cache_size: int,
    candidate_fn,
    rng: np.random.Generator | None = None,
):
    """
    Simulate a proactive (overnight) cache refresh per period with minimal moves.

    Parameters
    ----------
    all_results : list[pd.DataFrame]
        One DataFrame per period. Must include columns:
        - 'pnfsid' (file id)
        - 'true_label' (1 if accessed in this period, else 0)
        - 'period' (period id)
        Any other columns your candidate_fn needs (e.g., 'prediction').
    cache_size : int
        Number of files to keep in cache each period.
    candidate_fn : callable
        Signature: candidate_fn(df: pd.DataFrame, cache_size: int, rng: np.random.Generator | None) -> set[str]
        Returns the set of pnfsid candidates (size <= cache_size) for the given period.
        Examples: ML top-K by 'prediction', random choice, heuristics, etc.
    rng : np.random.Generator | None
        RNG for candidate_fn (optional).

    Returns
    -------
    period_history : pd.DataFrame
        Per-period metrics with columns:
        ['period','hits','requests','misses','moves','repeatness','precision_at_K','cum_hit_rate','cache_size_now']
    summary : dict
        Final totals and cumulative hit rate.
    """
    files_in_cache = pd.DataFrame([], columns=['pnfsid', 'period'])  # persistent state

    total_hits = 0
    total_requests = 0
    total_moves = 0
    total_repeats = 0

    period_logs = []
    j = 0
    for df in all_results:
        # if j>(190 -5):
        #     break
        current_period = df['period'].iloc[0]

        # 1) candidate set for this period (from your model / baseline)
        candidates = set(candidate_fn(df, cache_size, rng))

        # 2) ground truth accesses this period
        accessed = set(df.loc[df['y1'] == 1, 'pnfsid'])

        # 3) hit/miss using proactive cache = candidates (your operational model)
        hits = len(accessed & candidates)
        requests = len(accessed)
        misses = requests - hits

        # 4) movement accounting BEFORE updating cache
        in_cache = set(files_in_cache['pnfsid'])
        already_in = candidates & in_cache    # NOT moved (repeatness)
        to_insert = list(candidates - in_cache)  # moved

        repeatness = len(already_in)
        moves = len(to_insert)

        # 5) refresh recency for kept items
        if repeatness:
            files_in_cache.loc[
                files_in_cache['pnfsid'].isin(already_in), 'period'
            ] = current_period

        # 6) add without eviction if free capacity
        capacity = cache_size - len(in_cache)
        add_without_evict = max(0, min(len(to_insert), capacity))
        direct_add = to_insert[:add_without_evict]

        # 7) evict least-recent for remaining inserts
        remaining = len(to_insert) - add_without_evict
        if remaining > 0:
            victims = files_in_cache.sort_values('period', ascending=True).head(remaining)
            files_in_cache = files_in_cache.drop(victims.index)
            extra_add = to_insert[add_without_evict:add_without_evict + remaining]
        else:
            extra_add = []

        # 8) perform insertions (mark with current period)
        if direct_add or extra_add:
            new_rows = pd.DataFrame({'pnfsid': direct_add + extra_add, 'period': current_period})
            files_in_cache = pd.concat([files_in_cache, new_rows], ignore_index=True)

        # 9) safety: enforce capacity
        if len(files_in_cache) > cache_size:
            overflow = len(files_in_cache) - cache_size
            drop_more = files_in_cache.sort_values('period', ascending=True).head(overflow)
            files_in_cache = files_in_cache.drop(drop_more.index)

        # 10) update totals and log
        total_hits     += hits
        total_requests += requests
        total_moves    += moves
        total_repeats  += repeatness

        cum_hit_rate = (total_hits / total_requests) if total_requests > 0 else 0.0
        precision_at_K = (hits / cache_size) if cache_size > 0 else 0.0

        j += 1
        period_logs.append({
            'period': current_period,
            'hits': hits,
            'requests': requests,
            'misses': misses,
            'moves': moves,
            'repeatness': repeatness,
            'precision_at_K': precision_at_K,
            'cum_hit_rate': cum_hit_rate,
            'precicision_at_k': precision_at_K,
            'cache_size_now': len(files_in_cache),
        })

    period_history = pd.DataFrame(period_logs).sort_values('period').reset_index(drop=True)
    summary = {
        'cache_size': cache_size,
        'total_hits': total_hits,
        'total_requests': total_requests,
        'final_cum_hit_rate': (total_hits / total_requests) if total_requests > 0 else 0.0,
        'total_moves': total_moves,
        'total_repeatness': total_repeats,
        'avg_precision_at_k': period_history['precision_at_K'].mean()
    }
    return period_history, summary
```

```python
# 1) ML Top-K by 'prediction'
def topk_by_prediction(df: pd.DataFrame, cache_size: int, rng=None) -> set:
    return set(df.head(cache_size)['pnfsid'])

# 2) Random baseline (reproducible if rng provided)
def random_candidates(df: pd.DataFrame, cache_size: int, rng=None) -> set:
    ids = df['pnfsid'].to_numpy()
    if rng is None:
        rng = np.random.default_rng(42)
    k = min(cache_size, len(ids))
    return set(rng.choice(ids, size=k, replace=False))
```

```python
rng = np.random.default_rng(123) 

cache_sizes = list(range(25_000, 300_000, 25_000))
all_summaries = []

for K in cache_sizes:
    # ML model
    ph_ml, sm_ml = simulate_cache(all_results, K, topk_by_prediction, rng)
    sm_ml['policy'] = 'ML_topK'
    
    # Skip first period for moves/repeats stats
    moves = ph_ml['moves'].iloc[1:]
    repeats = ph_ml['repeatness'].iloc[1:]
    sm_ml['moves_mean'] = moves.mean()
    sm_ml['moves_std'] = moves.std()
    sm_ml['repeats_mean'] = repeats.mean()
    sm_ml['repeats_std'] = repeats.std()
    all_summaries.append(sm_ml)

    # Random
    ph_rand, sm_rand = simulate_cache(all_results, K, random_candidates, rng)
    sm_rand['policy'] = 'Random'

    moves = ph_rand['moves'].iloc[1:]
    repeats = ph_rand['repeatness'].iloc[1:]
    sm_rand['moves_mean'] = moves.mean()
    sm_rand['moves_std'] = moves.std()
    sm_rand['repeats_mean'] = repeats.mean()
    sm_rand['repeats_std'] = repeats.std()
    all_summaries.append(sm_rand)

    # Heuristic
    ph_heur, sm_heur = simulate_cache(all_lru_results, K, topk_by_prediction, rng)
    sm_heur['policy'] = 'Heuristic'

    moves = ph_heur['moves'].iloc[1:]
    repeats = ph_heur['repeatness'].iloc[1:]
    sm_heur['moves_mean'] = moves.mean()
    sm_heur['moves_std'] = moves.std()
    sm_heur['repeats_mean'] = repeats.mean()
    sm_heur['repeats_std'] = repeats.std()
    all_summaries.append(sm_heur)
    print(K)
    
    plt.figure(figsize=(10, 5))
    plt.plot(ph_ml['period'], ph_ml['cum_hit_rate'], color='blue', marker='o')
    plt.plot(ph_heur['period'], ph_heur['cum_hit_rate'], color='green', marker='s')
    plt.plot(ph_rand['period'], ph_rand['cum_hit_rate'], color='orange', linestyle='--', marker='x')
    plt.title('Cumulative Hit Rate Over Time')
    plt.xlabel('Period')
    plt.ylabel('Cumulative Hit Rate')
    plt.grid(True)
    plt.ylim([0,1])
    plt.show()

df_summary = pd.DataFrame(all_summaries)
```
