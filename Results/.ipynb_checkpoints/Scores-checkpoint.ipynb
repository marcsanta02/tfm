```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
scores = pd.read_csv("scoresN20N10N5.csv")
```

```python
# As I always appended the results of different models in the csv I kept track of where each model stopped the execution,
# that is the reason for the hardcoded numbers, they limit one model results with the next one.
frontier1 = int(365*2/20 - 5)
frontier2 = 92
frontier3 = 139
frontier4 = 241
frontier5 = 313
frontier6 = 358
frontier7 = 380

scoresN20 = scores.iloc[:frontier1]
scoresN10 = scores.iloc[frontier1:frontier2]
scoresN5 = scores.iloc[frontier2:frontier3]
scoresM15N5_3P = scores.iloc[frontier3:frontier4]
scoresM15N5_12P = scores.iloc[frontier4:frontier5]
scoresN5_v2 = scores.iloc[frontier5:frontier6]
scoresN5_v3 = scores.iloc[frontier6:frontier7]
scoresN5_v4 = scores.iloc[frontier7:]
```

```python
def format_mean_std(mean, std, precision=4):
    return f"{mean:.{precision}f} ± {std:.{precision}f}"

def weighted_mean_std_per_column(df, weight_col='number of files', exclude_cols=('period',)):
    cols = [
        col for col in df.select_dtypes(include=[np.number]).columns
        if col not in exclude_cols and col != weight_col
    ]
    weights = df[weight_col]
    
    data = {}
    for col in cols:
        values = df[col]
        mean = np.average(values, weights=weights)
        variance = np.average((values - mean) ** 2, weights=weights)
        std = np.sqrt(variance)
        data[col] = {'mean': mean, 'std': std}
        
    # Add ratios without stddev
    for metric in ["auc", "recall", "accuracy"]:
        num = data[f'unseen_{metric}']
        denom = data[f'val_{metric}']
        
        ratio = num['mean'] / denom['mean']
        ratio_std = ratio * np.sqrt(
            (num['std'] / num['mean'])**2 + (denom['std'] / denom['mean'])**2
        )
        
        data[f'ratio_{metric}'] = {'mean': ratio, 'std': ratio_std}
    
    return data

# Compute for both dataframes
summary1 = weighted_mean_std_per_column(scoresN10)
summary2 = weighted_mean_std_per_column(scoresN20)
summary3 = weighted_mean_std_per_column(scoresN5)
summary4 = weighted_mean_std_per_column(scoresM15N5_3P)
summary5 = weighted_mean_std_per_column(scoresM15N5_12P)
summary6 = weighted_mean_std_per_column(scoresN5_v2)
summary7 = weighted_mean_std_per_column(scoresN5_v3)
summary8 = weighted_mean_std_per_column(scoresN5_v4)



# Build the comparison DataFrame
rows = []
for metric in summary1.keys():
    m1 = summary1[metric]
    m2 = summary2[metric]
    m3 = summary3[metric]
    m4 = summary4[metric]
    m5 = summary5[metric]
    m6 = summary6[metric]
    m7 = summary7[metric]
    m8 = summary8[metric]




    

    val1 = format_mean_std(m1['mean'], m1['std'])
    val2 = format_mean_std(m2['mean'], m2['std'])
    val3 = format_mean_std(m3['mean'], m3['std'])
    val4 = format_mean_std(m4['mean'], m4['std'])
    val5 = format_mean_std(m5['mean'], m5['std'])
    val6 = format_mean_std(m6['mean'], m6['std'])
    val7 = format_mean_std(m7['mean'], m7['std'])
    val8 = format_mean_std(m8['mean'], m8['std'])






    rows.append({
        'metric': metric,
        'M=15 N=5 DF=12P': val5,
        'M=15 N=5 DF=3P': val4,
        'N=5': val3,
        'N=5 + Feature Engineering': val6,
        'N=5 + Feature Engineering + Tunning': val7,
        'N=5 + Feature Engineering + Recurrent Tunning': val8,
        'N=10': val1,
        'N=20': val2
    })

comparison_df = pd.DataFrame(rows).set_index('metric')

comparison_df
```

```python
all_params = []

for i in range(365*3 // 5 - 5*5):
    if i % 14 == 0:
        filename = f"v2_best_params_period_{i}_{i+13}.json"
        with open(filename, 'r') as file:
            data = json.load(file)
            all_params.append(data)
all_params = pd.DataFrame(all_params, index=[ i for i in range(365*3 // 5 - 5*5) if i%14==0] )
            
all_params            
        
```

```python
all_params_run0 = []

for i in range(365*3 // 5 - 11*5):
    if i % 14 == 0:
        filename = f"best_params_period_{i}_{i+13}.json"
        with open(filename, 'r') as file:
            data = json.load(file)
            all_params_run0.append(data)
all_params_run0 = pd.DataFrame(all_params_run0, index=[ i for i in range(365*3 // 5 - 11*5) if i%14==0] )
            
all_params_run0            
        
```

```python
# Define x-axis: indices where hyperparam search was performed
x = np.arange(0, len(all_params)) * 14  # assuming a search every 14 time periods
x2 = np.arange(0, len(all_params_run0)) * 14

param_ranges = {
    'max_depth': (3, 10),
    'learning_rate': (0.01, 0.3),
    'subsample': (0.5, 1.0),
    'colsample_bytree': (0.5, 1.0),
    'min_child_weight': (1, 10),
    'gamma': (0.0, 5.0),
    'reg_alpha': (0.0, 1.0),
    'reg_lambda': (0.5, 1.5),
    'early_stopping_rounds': (10, 30),
    'min_delta': (0.005, 0.1),
}

# Plot evolution of hyperparams
for col in all_params.columns:
    y = all_params[col].values
    y2 = all_params_run0[col].values
    mean = np.mean(y)
    std = np.std(y)
    mean_std_ratio = mean / std if std != 0 else np.nan

    plt.figure(figsize=(8, 4))
    plt.plot(x, y, marker='o', label="Search 1")
    plt.plot(x2, y2, marker='o', color='orange', label="Search 2")
    plt.title(f"{col} ($\mu$/$\sigma$ = {mean_std_ratio:.2f})")
    plt.xlabel("Time Period")
    plt.ylabel(col)
    plt.grid(True)

    # Set x-axis limits to show the full simulation range
    plt.xlim(x[0], x[-1])
    plt.ylim(param_ranges[col])
    plt.axhline(param_ranges[col][0], color='gray', linestyle='--', linewidth=0.5)
    plt.axhline(param_ranges[col][1], color='gray', linestyle='--', linewidth=0.5)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
